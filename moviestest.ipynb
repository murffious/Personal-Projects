```
import json
import simplejson
import requests
import urllib.parse


with open('/mymovies.json', 'r') as f:
    movies = json.load(f)
responses = list()

data = []
with open('/myfile.txt', "a") as f1:
   
    for movie in movies:
        # print(movie["movie_name"])
        query = movie["movie_name"]
        rr = requests.get(f'http://www.omdbapi.com/?apikey=60e45c02&t=${urllib.parse.quote(query)}')
        if 'Movie not found' in json.dumps(rr.json()):
          f1.write('{"Title": "' + movie["movie_name"]+'"}'+"\n")
          responses.append(movie["movie_name"]+" not found")
        else:  
          f1.write(rr.text+"\n")
          responses.append(json.dumps(rr.json()))
  
f1.close()
f.close()
print(responses)

```
